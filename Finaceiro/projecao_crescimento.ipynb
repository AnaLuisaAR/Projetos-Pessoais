```python
valor_inicial = 1400000
custo_inicial = 30000
inflacao = 0.05
rendimento = 0.1
aporte = 1000
tempo = 40
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ciclo(valor_inicial, custo_inicial, aporte, rendimento, inflacao):
    valor = valor_inicial
    custo = custo_inicial
    rend = rendimento/12
    inf = inflacao/12
    for _ in range(12):
        valor = valor*(1+rend) + aporte
        custo = custo*(1+inf)
    renda = valor*(rend-inf)
    return valor, custo, renda
```

```python
periodo = np.arange(0, tempo+1, 1)
valor = [valor_inicial/1000000]
custo = [custo_inicial]
renda = [valor_inicial*(rendimento-inflacao)/12]
```

```python
for i in range(tempo):
    v, c, r = ciclo(valor[i]*1000000, custo[i], aporte, rendimento, inflacao)
    valor.append(v/1000000)
    custo.append(c)
    renda.append(r)
    aporte *= 1.1
```

```python
fig = plt.figure()
ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx()
ax2.bar(periodo, valor, label = 'reserva (milh√µes)')
ax.plot(periodo, custo, label = 'custo')
ax.plot(periodo, renda, label = 'renda')

fig.legend()
plt.show()
```
